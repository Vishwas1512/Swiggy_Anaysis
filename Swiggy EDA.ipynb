# UniVarate

```python
plt.Figure(figsize=(8,10))
sns.boxplot(df['Price'])
plt.title('Boxplot of Price')
plt.show()
```

**Removing outliers**

```python
Price_Q1=df['Price'].quantile(0.25)
```

```python
Price_Q3=df['Price'].quantile(0.75)
```

```python
Price_IQR =Price_Q3-Price_Q1
```

```python
Price_IQR
```

```python
Price_lower=Price_Q1-1.5*Price_IQR
Price_higher=Price_Q3+1.5*Price_IQR
```

```python
df = df[(df['Price'] >= Price_lower) & (df['Price'] <= Price_higher)]
```

```python
plt.Figure(figsize=(8,10))
plt.boxplot(df['Price'])
plt.title('Boxplot of Price')
plt.show()
```

```python
plt.Figure(figsize=(8,10))
plt.boxplot(df['Avg ratings'])
plt.title('Boxplot of Avg ratings')
plt.show()
```

**Removing outliers**

```python
Total_ratings_Q1=df['Total ratings'].quantile(0.25)
```

```python
Total_ratings_Q3=df['Total ratings'].quantile(0.75)
```

```python
Total_ratings_IQR = Total_ratings_Q3-Total_ratings_Q1
```

```python
Total_ratings_lower = Total_ratings_Q1-1.5*Total_ratings_IQR
Total_ratings_higher = Total_ratings_Q3+1.5*Total_ratings_IQR
```

```python
df=df[(df['Total ratings']>=Total_ratings_lower) & (df['Total ratings']<=Total_ratings_higher)]
```

```python
plt.Figure(figsize=(8,10))
plt.boxplot(df['Total ratings'])
plt.title('Boxplot of Total ratings')
plt.show()
```

```python
plt.Figure(figsize=(8,10))
plt.boxplot(df['Delivery time'])
plt.title('Boxplot of Delivery time')
plt.show()
```

**Removing outliers**

```python
Delivery_time_Q1 = df['Delivery time'].quantile(0.25)
Delivery_time_Q3 = df['Delivery time'].quantile(0.75)
```

```python
Delivery_time_IQR=Delivery_time_Q3-Delivery_time_Q1
```

```python
Delivery_time_lower=Delivery_time_Q1-1.5*Delivery_time_IQR
Delivery_time_Higher=Delivery_time_Q3-1.5*Delivery_time_IQR
```

```python
df=df[(df['Delivery time'] >= Delivery_time_lower) & (df['Total ratings'] <= Delivery_time_Higher)]
```

```python
plt.Figure(figsize=(8,10))
plt.boxplot(df['Delivery time'])
plt.title('Boxplot of Delivery time')
plt.show()
```

* Customer Demographics by Area and City:
The data includes information on delivery times and orders placed in different areas and cities like Koramangala and Indiranagar in Bangalore. This insight allows Swiggy to analyze demand patterns and adjust operational capacity according to the area's delivery performance.
* Restaurant Pricing and Ratings:
Restaurants such as Tandoor Hut and Kim Lee have high average ratings (4.4), while New Punjabi Hotel has a slightly lower rating (3.9). Evaluating ratings alongside pricing can help in identifying high-performing, cost-effective restaurants that customers prefer.

* Restaurant Performance by Total Ratings:
Restaurants with a high number of total ratings, such as New Punjabi Hotel (500 ratings), indicate popularity and a large customer base. Monitoring customer feedback from these restaurants can provide Swiggy with critical insights to maintain service quality

* Delivery Time Analysis:
Delivery times vary across areas, with Indiranagar having longer average delivery times (63 minutes for Nh8) compared to other areas. Swiggy can focus on optimizing delivery routes or increasing the number of delivery personnel in such high-traffic areas to reduce delays.
